```python
distribution_df=data.groupby(['distribution_channel','hotel'])['adr'].mean().reset_index()
plt.figure(figsize=(12,16))
sn.barplot(x='distribution_channel',y='adr',data=distribution_df,hue='hotel')
plt.xlabel("Distribution Channel", fontsize = 16)
plt.ylabel("ADR", fontsize = 16)
plt.title('ADR across Distribution Channel', fontsize = 20)

# To show
plt.show()
     
```

```python
booking_by_month=data.groupby(['arrival_date_month'])['hotel'].count().reset_index().rename(columns={'hotel':'counts'})
months=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
booking_by_month['arrival_date_month']=pd.Categorical(booking_by_month['arrival_date_month'], categories = months, ordered = True)
booking_by_month = booking_by_month.sort_values('arrival_date_month')

booking_by_month
```

```python
plt.figure(figsize = (14,6))

# Plotting lineplot on x- months & y- bookings counts
sn.lineplot(x = booking_by_month['arrival_date_month'], y = booking_by_month['counts'])

# Set title
plt.title('Number of bookings across each month', fontsize = 20)

# Set labels
plt.xlabel('Month', fontsize = 16)
plt.ylabel('Number of bookings', fontsize = 16)

# To show
plt.show()
```
